```python
import numpy as np
import pandas as pd
```

```python
column_names = ['user_id', 'item_id', 'rating', 'timestamp']
df = pd.read_csv('users.data', sep='\t', names=column_names)
```

```python
df.head()
```

```python
len(df)
```

```python
movie_titles = pd.read_csv("movies.csv")
movie_titles.head()
```

```python
len(movie_titles)
```

```python
df = pd.merge(df, movie_titles, on='item_id')
df.head()
```

```python
moviemat = df.pivot_table(index='user_id',columns='title',values='rating')
moviemat.head()
```

Aramak istediğimiz filmi buraya yazıyoruz.<br>
Örn: "Seven (Se7en) (1995)", "Forrest Gump (1994)", "Terminator, The (1984)", "Jurassic Park (1993)", "Lion King, The (1994)"

```python
movies_user_ratings = moviemat["Terminator, The (1984)"]
movies_user_ratings.head()
```

```python
similar_to_movies = moviemat.corrwith(movies_user_ratings)
```

```python
similar_to_movies
```

```python
corr_movies = pd.DataFrame(similar_to_movies, columns=['Correlation'])
corr_movies.dropna(inplace=True)
corr_movies.head()
```

```python
corr_movies.sort_values('Correlation',ascending=False).head(10)
```

```python
df.head()
```

```python
df.drop(['timestamp'], axis = 1)
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings.sort_values('rating',ascending=False).head()
```

```python
ratings['rating_vote_num'] = pd.DataFrame(df.groupby('title')['rating'].count())
ratings.head()
```

```python
ratings.sort_values('rating_vote_num',ascending=False).head()
```

```python
corr_movies.sort_values('Correlation',ascending=False).head(10)
```

```python
corr_movies = corr_movies.join(ratings['rating_vote_num'])
corr_movies.head()
```

```python
corr_movies[corr_movies['rating_vote_num']>120].sort_values('Correlation',ascending=False).head()
```
